```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, figsize=(10, 10))
ax.axis("off")

# Plot the GeoDataFrame without edgecolor initially
fig = merged.plot(
    column="MURDER",
    cmap="RdPu",
    linewidth=0.5,
    ax=ax,
    legend=True,
)

# Iterate through the patches and set the edgecolor
for patch in fig.collections:
    patch.set_edgecolor("0.2")  # Set edgecolor for each patch

fig.set_title("Statewise Murder Cases Reported(2000 - 2012)", size=17)
plt.show()
```
